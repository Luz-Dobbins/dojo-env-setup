 Backticks
Occasionally a column name will require backticks to help SQL understand that we are referring to a column name that is a bit unusual. This happens when:

---a column name happens to also be a SQL keyword (select, from, order, value, range)

---the column name has spaces

For example, some of the columns in the school_details table have spaces in the name. One option would be to avoid naming columns this way in the first place! But, if you are working with a table in a database that you are not able to change, you can still access these columns by surrounding the column name with backticks (').

Be careful, this is NOT an apostrophe or single quote! On most English keyboards, the backtick button shares a key with the tilde (~) to the left of the 1 key on your keyboard.

For example, let's SELECT the CRCDistance (in Km) column from the school_details table. Note that the actual column name CRCDistance (in Km) contains spaces. This will cause an error if we do not tell SQL that those spaces are actually part of the column name, so we write the column like this: `CRCDistance (in Km)`

```python
# Write the SQL query
sql = """SELECT `CRCDistance (in Km)`
FROM school_details;
"""
#Execute the query
pd.read_sql_query(sql, engine)
```

WHERE

So far we have selected entire columns of data, but oftentimes, you will want to filter your results based on some criteria. The WHERE keyword comes in handy by allowing us to specify conditions for which data will be included. WHERE allows us to filter the output.

We can filter on numerical values using these operators:

-- Less than: <
    
-- Greater than: >
    
-- Less than or equal to: <=
    
-- Greater than or equal to: >=
    
-- Equal to: =
    
-- Not equal to: <> or !=

```python
sql = """
SELECT EstablishmentName, CRCName, `CRCDistance (in Km)`
FROM school_details
WHERE `CRCDistance (in Km)` > 5;
"""

pd.read_sql_query(sql, engine)
```

```python
sql = """
SELECT EstablishmentName
From school_details
WHERE SchoolLevel = 'PS';
"""

pd.read_sql_query(sql, engine)
```

Wildcards (%)

Wildcards (%) can replace text when you don't know the exact text that you need to match. For example, we could pull all data that start or end with a certain letter or letters. To do this, just replace the unknown or "wildcard" text with a "%" symbol.

Since we are not exactly matching text, we no longer want to use an equal sign. We use the keyword LIKE in this case.

```python
sql = """SELECT EstablishmentName, PostofficeName
FROM school_details
WHERE EstablishmentName LIKE 'New%'
"""
pd.read_sql_query(sql, engine)
```

```python
sql = """SELECT EstablishmentName
FROM school_details
WHERE EstablishmentName LIKE '%academy%'
"""
pd.read_sql_query(sql, engine)
```

OR and AND Statements
We can use OR and AND statements to further refine our WHERE filter

---AND means that both of the conditionals have to be true in order to grab that data.

---OR means that only one of them has to be true in order to be included.

```python
sql = """SELECT EstablishmentName
FROM school_details
WHERE EstablishmentName LIKE 'new%' AND EstablishmentName LIKE'%academy%';
"""
pd.read_sql_query(sql, engine)
```

```python
sql = """SELECT EstablishmentName
FROM school_details
WHERE EstablishmentName LIKE 'new%' OR EstablishmentName LIKE'%academy%';
"""
pd.read_sql_query(sql, engine)
```

Multiple Columns

We can also filter on multiple columns within the same WHERE statement using the AND and OR operators.

```python
sql = """SELECT EstablishmentName
FROM school_details
WHERE EstablishmentName LIKE 'new%' AND `CRCDistance (in Km)` > 5;
"""
pd.read_sql_query(sql, engine)
```

```python
sql = """SELECT EstablishmentName
FROM school_details
WHERE EstablishmentName LIKE 'new%' AND EstablishmentName LIKE'%academy%' AND `CRCDistance (in Km)` > 5;
"""
pd.read_sql_query(sql, engine)
```

ORDER BY

The ORDER BY keyword is used to sort your results by a column. This works with both numeric and non-numeric columns. By default, your results will be ordered in ascending order (lowest value to highest value) for numeric columns and alphabetically for non-numeric columns.

ORDER BY Using DESC

The ASC and DESC keywords can be used in conjunction with the ORDER BY keyword to specify what order you want your results displayed in. ASC stands for 'ascending' and it sorts results lowest to highest or alphabetically. ASC is unnecessary since it is the default sorting for ORDER BY, but can improve readability. DESC stands for 'descending' and can be used to sort results from highest to lowest or in reverse alphabetical order.

ORDER BY With WHERE Statement

You can order by results that have been filtered with a WHERE statement:

ORDER BY Multiple Columns

If you have multiple columns to sort on, you can ORDER BY multiple columns. Just separate the columns you want to sort by with a comma and the results will be sorted by the first column you order by followed by the second, and so on for however many columns you are sorting by.

LIMIT

If we do not need to grab all of the rows from the table we are querying from, we can limit the results to a certain number of rows. This is useful for only getting the information you need and for speed and efficiency. This is similar to the .head(n) method in Pandas where you can display n rows in your dataframe! LIMIT is going to be the last keyword in your query.



```python
sql = """
SELECT EstablishmentName, `CRCDistance (in Km)`, CRCName 
FROM school_details
ORDER BY `CRCDistance (in Km)` DESC;
"""
pd.read_sql_query(sql, engine)
```

```python
sql = """
SELECT EstablishmentName, `CRCDistance (in Km)`, CRCName 
FROM school_details
ORDER BY `CRCDistance (in Km)` DESC
LIMIT 1;
"""
pd.read_sql_query(sql, engine)
```

```python
sql = """SELECT EstablishmentName
FROM school_details
WHERE EstablishmentName LIKE '%government%'
LIMIT 10;
"""
pd.read_sql_query(sql, engine)
```

```python
sql = """SELECT EstablishmentName, CRCName
FROM school_details
ORDER BY CRCName, EstablishmentName
LIMIT 20;
"""
pd.read_sql_query(sql, engine)
```

Aggregates

Aggregates are often used in conjunction with grouping if we want to get aggregate statistics on different groups.

```python
#Find the average DistrictHQDistance (in Km) when schools are grouped by LocatedinRuralOrUrban
sql = """
SELECT LocatedInRuralOrUrban, AVG(`DistrictHQDistance (in Km)`)
FROM school_details
GROUP BY LocatedInRuralOrUrban;
"""
pd.read_sql_query(sql, engine)
```

```python
#Find the maximum DistrictHQDistance when results are grouped by PostOfficeName
sql = """SELECT PostOfficeName, MAX(`DistrictHQDistance (in Km)`)
FROM school_details
GROUP BY PostOfficeName;
"""
pd.read_sql_query(sql, engine)
```

```python
#Find the number of schools in each CRC and order them by high to low
sql = """SELECT CRCName, COUNT(*)
FROM school_details
GROUP BY CRCName
ORDER BY COUNT(*) DESC;
"""
pd.read_sql_query(sql, engine)
```

The above code works, but this is a great example of when naming the aggregate value with AS can be helpful. Notice below that the COUNT(*) is named AS "NumberofSchools". This name can then be used in the ORDER BY statement and also shows up in the output.

```python
sql = """SELECT CRCName, COUNT(*) AS NumberofSchools
FROM school_details
GROUP BY CRCName
ORDER BY NumberofSchools DESC;
"""
pd.read_sql_query(sql, engine)
```

HAVING

To filter values based on the results of aggregate functions with GROUP BY, the HAVING keyword is used. This is very similar to the WHERE keyword, but HAVING must be used with GROUP BY.

```python
sql = """SELECT CRCName, COUNT(*)
FROM school_details
GROUP BY CRCName
HAVING COUNT(*) > 20;
"""
pd.read_sql_query(sql, engine)
```

```python
sql = """SELECT CRCName, COUNT(*) AS NumberofSchools
FROM school_details
GROUP BY CRCName
HAVING NumberofSchools > 20;
"""
pd.read_sql_query(sql, engine)
```
